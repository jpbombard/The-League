## ESPN Data 

```python
from espn_api.football import League
from dotenv import load_dotenv
import os
import pandas as pd

load_dotenv()
espn_s2 = os.getenv("ESPN_S2")
swid = os.getenv("SWID")

LEAGUE_ID = 48346
SEASON_YEAR = 2024

league = League(
    league_id=LEAGUE_ID,
    year=SEASON_YEAR,
    espn_s2=espn_s2,
    swid=swid
)

data = []
for team in league.teams:
    data.append({
        "Team Name": team.team_name,
        "Owner": team.owners[0] if team.owners else "Unknown",
        "Wins": team.wins,
        "Losses": team.losses,
        "Points For": team.points_for,
        "Points Against": team.points_against
    })

df = pd.DataFrame(data)

df
```
